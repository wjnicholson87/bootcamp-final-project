```python
def win_count(bracket):
    wins = {}
    win_count = 0
    
    bracket_sets = smash.bracket_show_sets(bracket, 1)
    bracket_df = pd.DataFrame(data=bracket_sets, columns=bracket_sets[0].keys())
    players_list = bracket_df.loc[:, ['entrant1Name', 'entrant2Name']]
    unique_players = np.unique(players_list)
    
    for player in unique_players:
        win_count = 0
        for i in range(0, len(bracket_df['winnerName'])):
            if player == bracket_df['winnerName'][i]:
                win_count += 1
                print(bracket_df['winnerName'][i])
        wins[player] = win_count
        
    return wins

win_count(1829817)
```
